_(5 Points)_. Propose a two-parameter model, something simple, like an exponential, i.e., $y = e^ax + xb$ or a quadratic polynomial, i.e., $y = ax + bx^2$ , would work. Simulate $N = 100$ data points with your model and add some Gaussian noise to mimic a measurement of real data. Just like we did in class.

```python
model = lambda x,a,b: a*x + b*(x**2)
a = 11
b = np.pi
N=100
dom = np.linspace(0, 100,N)
y_real = model(dom,a,b)
plt.plot(dom, y_real)
```

```python
y_syn = y_real + np.random.rand(N)
y_syn = y_real + y_syn * 0.2*np.random.randn(N)

plt.plot(y_real)
plt.scatter(dom, y_syn)
```
