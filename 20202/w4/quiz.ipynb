```python
import numpy as np
import sympy as sy
from sympy.functions import sin,cos
import matplotlib.pyplot as plt
```

```python
# Factorial function
def factorial(n):
    if n <= 0:
        return 1
    else:
        return n*factorial(n-1)

# Taylor approximation at x0 of the function 'function'
def taylor(function,x0,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (function.diff(x,i).subs(x,x0))/(factorial(i))*(x-x0)**i
        i += 1
    return p

x = sy.Symbol('x')
f = sin(x) + cos(x)

x_real = np.linspace(0, np.pi, 1000)

y_real = np.sin(x_real) + np.cos(x_real)

# Taylor series expansion at n=1
y_1 = taylor(f, 0, 1)
y_1_d = [y_1.subs(x, q) for q in x_real]
# Taylor series expansion at n=2
y_2 = taylor(f, 0, 2).subs(x, x_real)
y_2_d = [y_2.subs(x, q) for q in x_real]
# Taylor series expansion at n=3
y_3 = taylor(f, 0, 3).subs(x, x_real)
y_3_d = [y_3.subs(x, q) for q in x_real]

plt.plot(x_real, y_real, label="real f(x)")
plt.plot(x_real, y_1_d, label="Taylor Series n=1")
plt.plot(x_real, y_2_d, label="Taylor Series n=2")
plt.plot(x_real, y_3_d, label="Taylor Series n=3")

plt.legend()
```
