```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform
%matplotlib inline
```

```python
N = 100000

times=N
estimates = np.zeros(N)
estimates[0] = -5

for i in range(times-1):
    sample_u = np.random.uniform(0,1)
    sample_p = np.random.normal(estimates[i])
    if sample_u < min(1, norm.pdf(sample_p) / norm.pdf(estimates[i])):
        estimates[i+1] = sample_p
    else:
        estimates[i+1] = estimates[i]
        
plt.hist(estimates, density=True, bins=int(N/100))
dom_1 = np.linspace(-3, 3, N)
ran_1 = norm.pdf(dom_1)
plt.plot(dom_1, ran_1, 'r')
plt.show()
```

```python
random_numbers = np.random.normal(size=N)
plt.hist(random_numbers, density=True, bins=int(N/100))
dom_2 = np.linspace(-3, 3, N)
ran_2 = norm.pdf(dom_2)
plt.plot(dom_1, ran_2, 'r')
plt.show()
```
